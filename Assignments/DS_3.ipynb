### Q.1 Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a carâ€™s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.
```
Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]
```


````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
days = [1, 2, 3, 4, 5, 6, 7, 8]
speed = [60, 62, 61, 58, 56, 57, 46, 63]

plt.plot(days, speed)
plt.xlabel("Days")
plt.ylabel("Speed")
plt.title("Car Speed Measurement")
plt.show()
````


### 2. Now to above car data apply some string formats  like line style example green dotted line, marker shape like +, change markersize, markerface color etc.



````python
f, axis = plt.subplots(2, 3, figsize=(20, 15))

days = [1, 2, 3, 4, 5, 6, 7, 8]
speed = [60, 62, 61, 58, 56, 57, 46, 63]

plt.suptitle("Car Speed Measurement", fontsize=0)

axis[0, 0].set_title("Green doted line",fontsize=20)
axis[0, 0].set_xlabel("Days", fontsize=15)
axis[0, 0].set_ylabel("Speed", fontsize=15)
axis[0, 0].plot(days, speed, ":g")

axis[0, 1].set_title("Marker shape +",fontsize=20)
axis[0, 1].set_xlabel("Days", fontsize=15)
axis[0, 1].set_ylabel("Speed", fontsize=15)
axis[0, 1].plot(days, speed, "+")

axis[0, 2].set_title("Marker size 6",fontsize=20)
axis[0, 2].set_xlabel("Days", fontsize=15)
axis[0, 2].set_ylabel("Speed", fontsize=15)
axis[0, 2].plot(days, speed, "^-m", ms=6)

axis[1, 0].set_title("Red marker face color",fontsize=20)
axis[1, 0].set_xlabel("Days", fontsize=15)
axis[1, 0].set_ylabel("Speed", fontsize=15)
axis[1, 0].plot(days, speed, "o-k", mfc="r")

axis[1, 1].set_title("blue marker edge color",fontsize=20)
axis[1, 1].set_xlabel("Days", fontsize=15)
axis[1, 1].set_ylabel("Speed", fontsize=15)
axis[1, 1].plot(days, speed, "o-.c", mec="b")

axis[1, 2].set_title("Custom color",fontsize=20)
axis[1, 2].set_xlabel("Days", fontsize=15)
axis[1, 2].set_ylabel("Speed", fontsize=15)
axis[1, 2].plot(days, speed, "*--y", mec="#f2342c")

plt.show()
````


### 4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.


````python
import matplotlib.pyplot as plt
import numpy as np
import math

X = np.arange(0, math.pi*4, 0.05)
Y1 = np.sin(X)
Y2 = np.cos(X)

plt.plot(X, Y1)
plt.plot(X, Y2)

plt.xlabel('x values from 0 to 4pi')  
plt.ylabel('sin(x) and cos(x)')

plt.legend(["sin(x)","cos(x)"])
plt.show()
````


### 5. Plot Simple bar chart showing popularity of Programming Languages.
```
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]
```
### Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.


````python
languages = ['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [56, 39, 34, 34, 29]
security = [44, 36, 55, 50, 42]

x_axis = np.arange(len(languages))

plt.bar(x_axis-0.2, popularity, 0.4, label="Popularity")
plt.bar(x_axis+0.2, security, 0.4,color='m', label="Security")
plt.xticks(x_axis, languages)

plt.xlabel("Languages")
plt.ylabel("Popularity and Security")
plt.title("Popularity of major Programming Languages")
plt.legend()
plt.show()
````

````python
# Horizontal graph
plt.barh(x_axis-0.2, popularity, 0.4, label="Popularity")
plt.barh(x_axis+0.2, security, 0.4,color='m', label="Security")
plt.yticks(x_axis, languages)

plt.ylabel("Languages")
plt.xlabel("Popularity and Security")
plt.title("Popularity of major Programming Languages")
plt.legend()

plt.show()
````

```
marks = [ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]
40-60: Below Average
60-80: Average
80-100: Excellent
```

### Histogram showing Below Average, Average and Execellent distribution


````python
marks = np.array([61, 86, 42, 46, 73, 95, 65, 78, 53, 92, 55, 69, 70, 49, 72, 86, 64])
below_average = marks[np.logical_and(marks >= 40, marks < 60)]
average = marks[np.logical_and(marks >= 60, marks < 80)]
exellect = marks[np.logical_and(marks >= 80, marks < 100)]

plt.hist(marks,ec='k')
plt.xlabel('Bins')
plt.ylabel('Marks')
plt.title('Marks of students')
plt.vlines([40, 60, 80, 100], ymin=0, ymax=4, colors='k', linestyles='--')
plt.show()
````

````python
bins = [40, 60, 80, 100]
plt.hist(marks, bins, color='#f57c00', ec='b')
plt.xlabel('Bins')
plt.ylabel('Marks')
plt.title('Histogram of marks')
plt.show()
````

````python
male = df[df["sex"] == "male"]
female = df[df["sex"] == "female"]
male_count, female_count = male.sex.value_counts(), female.sex.value_counts()

plt.scatter(male.fare, male.age, label="Male", marker="*", color="c")
plt.scatter(female.fare, female.age, label="Female", marker="^", color="r")
plt.title("Fare Paid and Age with respect to Gender")
plt.xlabel("Fare")
plt.ylabel("Age")
plt.legend()
plt.show()
````
