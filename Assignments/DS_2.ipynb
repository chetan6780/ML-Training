# Pandas - Data Analysis of IMDB movies data.



### Load the IMDb Dataset and read.


```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data/IMDB-Movie-Data.csv")
df
```

### View the dataset.


```python
print(df.head())
print(df.tail())
```

### Understand some basic information about the dataset and Inspect the dataframe Inspect the dataframe's columns, shapes, variable types etc.


```python
print("Information about the dataset:")
print("\nShape:",df.shape)
print("\nDimensions:",df.ndim)
print("\nLength:",len(df))
print("\nSize:",df.size)
print("\nColumns:",df.columns)
print("\nVariable types:",df.dtypes)
```


### Data Selection – Indexing and Slicing data


```python
# Dataframe.loc[ ] : Used for labels.
df.loc[[3,6,9]] #indexing
```

```python
df.loc[30:40,["Rank","Title","Genre","Rating"]]#slicing
```

```python
# Dataframe.iloc[ ] : Used for positions or integer based.
df.iloc[[3,6,9]] #indexing
```

```python
df.iloc[30:40,[0,1,2,8]] #slicing 
```

### Data Selection – Based on Conditional filtering


```python
high_rated = df[df["Rating"] >= 8.5]
print("All the highrated movies:\n")
high_rated
```

### Groupby operations


```python
dt = df.groupby("Rank")["Title"]
print(dt.first())
```

### Sorting operation.

```python
dt = df.sort_values('Rating')
print(dt["Rating"])
```

### Dealing with missing values


```python
null_data = df.isnull()
print(null_data.shape)
```

```python
# Filling null values with a single value.
df.fillna(0)
```

```python
# Filling null values with the previous ones
df.fillna(method='pad')
```

### Dropping columns and null values


```python
drop_null = df.dropna()
print(drop_null.shape)
```

### Apply() functions.

```python
dt = df.Rating.apply(lambda x: x*10)
dt
```
